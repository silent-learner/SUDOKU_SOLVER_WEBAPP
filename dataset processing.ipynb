```python
import numpy as np 
import pandas as pd
import os
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
BASE_DIR = './digits'
```

```python
os.listdir(BASE_DIR)
```

```python
images = []
labels = []

for folder in os.listdir(BASE_DIR):
    for img in os.listdir(os.path.join(BASE_DIR,folder)):
        mat = cv.imread(os.path.join(os.path.join(BASE_DIR,folder),img),cv.IMREAD_GRAYSCALE)
        images.append(mat)
        labels.append(int(folder))
```

```python
X = np.array(images)
y = np.array(labels)
```

```python
X.shape
```

```python
np.save('images.npy',X)
np.save('labels.npy',y)
```
